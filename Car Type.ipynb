```python
import pandas as pd

carsdata = pd.read_csv("fuel.csv")
carsdata
```

```python
carsdata['class_label'] = carsdata["drive"].apply(lambda x: 1 if x == 'Front-Wheel Drive' else 0 )
carsdata['class_label']
```

```python
carsdata
```

```python
cars_new = carsdata[["make", "model", "VClass", "drive", "displ", "comb", "class_label"]]

cars_new.head(15)
```

```python
from sklearn.model_selection import train_test_split
X = carsdata[["displ", "comb"]]
y = carsdata["class_label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# for educational purposes i will use StandardScaler to show something
#Ù„Ø£ØºØ±Ø§Ø¶ ØªØ¹Ù„ÙŠÙ…ÙŠØ© Ø³Ø£Ø³ØªØ®Ø¯Ù… StandardScaler Ù„Ø¥Ø¸Ù‡Ø§Ø± Ø´ÙŠØ¡ Ù…Ø§


from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = cars_new[['displ', 'comb']]
X_scaled = sc.fit_transform(X)
X_train_scaled, X_test_scaled, _, _ =train_test_split(X_scaled, y , random_state= 0 )
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train.values, y_train)
model_scaled = LogisticRegression().fit(X_train_scaled, y_train)
```

```python
y_pred=model.predict(X_test.values)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)

accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy score:", accuracy)

precision = metrics.precision_score(y_test,y_pred)
print("Precision score:", precision)

recall = metrics.recall_score(y_test,y_pred)
print("Recall score:", recall)
```

```python
#now training the model on the StandardScaled data
y_pred_scaled = model_scaled.predict(X_test_scaled)
```

```python
cm = metrics.confusion_matrix(y_test,y_pred_scaled)
print(cm)
accuracy = metrics.accuracy_score(y_test, y_pred_scaled)
print("Accuracy score sclaed:", accuracy)
precision = metrics.precision_score(y_test, y_pred_scaled)
print("Precision score scaled:", precision)
recall = metrics.recall_score(y_test, y_pred_scaled)
print("Recall score scaled:", recall)
```

```python
#the reason it show the same accuracy score scaled show the same result as normal accuracy
#is because numbers of data are close to each other and not far 
#we use StandardScaler only whe the numbers of data are far from eachother and we want to scale it between 0 and 1
```
